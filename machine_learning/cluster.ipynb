```python
import pandas as pd
import pyarrow.parquet as pq
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import boto3
from botocore.exceptions import NoCredentialsError
from minio import Minio
```

```python
# Set your MinIO server credentials
minio_endpoint = "http://localhost:9000"
minio_access_key = "minioadmin"
minio_secret_key = "minioadmin"
minio_bucket_name = "parquetfiles"
minio_file_name = "heart_attack_prediction_dataset.csv.parquet"

# Download the Parquet file from MinIO
def download_file_from_minio():
    try:
        s3 = boto3.client(
            "s3",
            endpoint_url=minio_endpoint,
            aws_access_key_id=minio_access_key,
            aws_secret_access_key=minio_secret_key,
        )

        with open(minio_file_name, "wb") as f:
            s3.download_fileobj(minio_bucket_name, minio_file_name, f)

    except NoCredentialsError:
        print("Credentials not available")
```

```python
table = pq.read_table(minio_file_name)
df = table.to_pandas()
df
```

```python
# Menggunakan kolom 'Exercise Hours Per Week' dan 'BMI' sebagai fitur clustering
features = df[['Exercise Hours Per Week', 'BMI']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
wcss = []  # Within-Cluster-Sum-of-Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)

# Plot metode elbow
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
k = 4  # Sesuaikan dengan jumlah cluster yang ditemukan melalui metode elbow
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(features_scaled)
```

```python
plt.scatter(features_scaled[:, 0], features_scaled[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('Hasil Clustering')
plt.xlabel('Exercise Hours Per Week (Scaled)')
plt.ylabel('BMI (Scaled)')
plt.legend()
plt.show()
```
